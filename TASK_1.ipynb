```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
df = pd.read_csv(url)

# Filter for year 2021
df_2021 = df[df['Year'] == 2021]

# Get top 10 countries by population
top10 = df_2021.sort_values(by='Value', ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(top10['Country Name'], top10['Value'] / 1e6)
plt.title("Top 10 Most Populated Countries (2021)")
plt.xlabel("Country")
plt.ylabel("Population (Millions)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load population dataset
url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
df = pd.read_csv(url)

# Filter for 2021 data
df_2021 = df[df['Year'] == 2021]

# Get top 10 most populated countries
top10 = df_2021.sort_values(by='Value', ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
plt.bar(top10['Country Name'], top10['Value'] / 1e6, color='purple')
plt.title("Top 10 Most Populated Countries (2021)")
plt.xlabel("Country")
plt.ylabel("Population (Millions)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate ages of 1000 people
np.random.seed(0)
ages = np.random.normal(loc=35, scale=10, size=1000)  # Mean 35, SD 10

# Plot histogram
plt.figure(figsize=(10,6))
plt.hist(ages, bins=20, color='skyblue', edgecolor='black')
plt.title("Simulated Age Distribution in a Population")
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.grid(True)
plt.show()
```
